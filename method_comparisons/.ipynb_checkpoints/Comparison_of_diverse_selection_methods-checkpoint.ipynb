```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10,7]
```

```python
coverage_testing = pd.read_csv('coverage_testing1.csv')
```

```python
coverage_testing['average_coverage'] = coverage_testing['Donor,Acceptor,NegIonizable,PosIonizable,Aromatic,Hydrophobe'.split(',')].mean(axis = 1)
```

```python
fig = sns.lineplot(data = coverage_testing, x='n_select', y = 'average_coverage', hue = 'method', style = 'use_coverage')
```

```python
coverage_testing
```

```python
fig = sns.lineplot(data = coverage_testing, x='n_select', y = 'average_coverage', hue = 'method')
```

```python
pivot_coverage_testing = pd.melt(coverage_testing, id_vars = ['n_select','method','use_coverage','transform_func'], var_name = 'pharmacophore', value_name = 'coverage')
```

```python
pivot_coverage_testing
```

```python
sns.set(font_scale=1.2)
sns.set_style("darkgrid")

fig = sns.lineplot(data = pivot_coverage_testing[pivot_coverage_testing['method'] == 'som'], x='n_select', y = 'coverage', hue = 'use_coverage')
plt.axvline(x=240, color = "pink", linestyle = "dashed")
```

```python
sns.set(font_scale=1.2)
sns.set_style("darkgrid")

fig = sns.lineplot(data = pivot_coverage_testing[pivot_coverage_testing['method'] == 'som'], x='n_select', y = 'coverage', hue = 'transform_func')
plt.axvline(x=240, color = "pink", linestyle = "dashed")
```
